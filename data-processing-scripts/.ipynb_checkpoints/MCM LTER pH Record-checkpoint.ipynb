# MCM LTER ALMMS Data Processing

This script was created by Jared Collins in 2025 to be used by the MCM LTER stream team as an easy way to organize data from the Active Layer & Moat Monitoring Stations (ALMMS), which were set up as part the Soil-Lake Inundation Moat Experiment (SLIME). The stations measuring volumetric soil content (VWC), conductivity, and temperature along the lake moats in both the soils and the water column, as well as air temperature and photosynthetically active radiation (PAR).

## Importing Packages + Setting Directory

It is very important that the computer being used for processing contains the proper packages, paths, and names for the data files that will be used. The hyporheic data is a little bit more difficult to work with than data processed in Aquarius- unlike data from Aquarius, it is not automatically converted into a dataframe from the .dat file.

**To use this script, please copy the .dat file you would like to convert to your downloads folder.**

Note: The datafiles from the CR1000X all have the same name with no date differentiation, so this can only be done using one data file per stream at a time.

```python
from datetime import date
import pandas as pd
import os
from pathlib import Path
import re
import numpy as np
```

```python
home = Path.home()
downloads_path = home / "Downloads"
downloads_path = downloads_path.as_posix()
print("Downloads folder:", downloads_path)
```

```python
slimeid_map = {
    'NFRX': 'NRFX',
    'SFRX': 'SFRX',
    'NELB': 'NELB',
    'SELB': 'SELB'}

malmms_map = {
    '1': 'h1_25cm',
    '2': 'h1_15cm',
    '3': 'h1_05cm',
    '4': 'h2_25cm',
    '5': 'h2_15cm',
    '6': 'h2_05cm',
    '7': 'h3_25cm',
    '8': 'h3_15cm',
    '9': 'h3_05cm',
    '10': 'h4_25cm',
    '11': 'h4_15cm',
    '12': 'h4_05cm'}

selb_map = {
    '1': 'h1_40cm',
    '2': 'h1_35cm',
    '3': 'h1_25cm',
    '4': 'h1_15cm',
    '5': 'h1_05cm',
    '6': 'h2_45cm',
    '7': 'h2_35cm',
    '8': 'h2_25cm',
    '9': 'h2_15cm',
    '10': 'h2_05cm',
    '11': 'h3_45cm',
    '12': 'h3_35cm',
    '13': 'h3_25cm',
    '14': 'h3_15cm',
    '15': 'h3_05cm',
    '16': 'h4_45cm',
    '17': 'h4_35cm',
    '18': 'h4_25cm',
    '19': 'h4_15cm',
    '20': 'h4_05cm'}
```

In order for this script to work properly, you will need to input necessary information about the files that you are hoping to publish:
- SLIME ID: NFRX, SFRX, NELB, SELB
- Season: The two years of the season (for example, 2024-2025 season is written as 2425)

**Edit the following cell with the information above. If done correctly, this is the only cell you will have to edit.**

```python
slimeid = 'SELB'
season = '2425'
```

```python
bur = pd.read_csv(f'{downloads_path}/{slimeid_map[slimeid]}_ALMMS (cr1000x)_SLIME_{slimeid}_BURIED.dat', skiprows = [0,2,3], sep = ',', low_memory = False)
exp = pd.read_csv(f'{downloads_path}/{slimeid_map[slimeid]}_ALMMS (cr1000x)_SLIME_{slimeid}_EXPOSED.dat', skiprows = [0,2,3], sep = ',', low_memory = False)
slimeid_lc = slimeid.lower()

def rename_ecp_columns(df: pd.DataFrame, slimeid: str) -> pd.DataFrame:
    '''Rename ECp(#) columns according to site mapping.'''
    site_code = slimeid_map.get(slimeid)
    mapping = selb_map if site_code == 'SELB' else malmms_map
    rename_dict = {}
    for col in df.columns:
        match = re.search(r'ECp\((\d+)\)', col)
        if match:
            ch = match.group(1)
            if ch in mapping:
                rename_dict[col] = f'soilec_{mapping[ch]}_dsm'
    return df.rename(columns=rename_dict)
    
def rename_temp_columns(df: pd.DataFrame, slimeid: str) -> pd.DataFrame:
    site_code = slimeid_map.get(slimeid)
    mapping = selb_map if site_code == 'SELB' else malmms_map
    rename_dict = {}
    for col in df.columns:
        match = re.search(r'Temp\((\d+)\)', col)
        if match:
            ch = match.group(1)
            if ch in mapping:
                rename_dict[col] = f'soilt_{mapping[ch]}_degc'
    if site_code == 'SELB':
        tc_depth_pattern = re.compile(r'selbbur_TC\d+_H(\d+)_(\d+)cm', re.IGNORECASE)
        tc_surface_pattern = re.compile(r'selbbur_TC\d+_H(\d+)_surface', re.IGNORECASE)
    else:
        prefix = slimeid[:2].lower()
        tc_depth_pattern = re.compile(rf'{prefix}bur_TC\d+_H(\d+)_(\d+)cm', re.IGNORECASE)
        tc_surface_pattern = re.compile(rf'{prefix}bur_TC\d+_H(\d+)_surface', re.IGNORECASE)
    for col in df.columns:
        match = tc_depth_pattern.search(col)
        if match:
            hole, depth = match.groups()
            rename_dict[col] = f'soilt_h{hole}_{depth}cm_degc'
    for col in df.columns:
        match = tc_surface_pattern.search(col)
        if match:
            hole = match.group(1)
            rename_dict[col] = f'soilt_h{hole}_surface_degc'
    return df.rename(columns=rename_dict)

def parse_hole_depth_from_col(col):
    m = re.search(r'h(\d+)_([0-9]+)cm', col)
    if m:
        return int(m.group(1)), int(m.group(2))
    return 99, 0

def sort_soil_columns(df: pd.DataFrame) -> list:
    cols = [c for c in df.columns if c.startswith('soilt_')]
    def sort_key(col):
        m_depth = re.search(r'h(\d+)_([0-9]+)cm', col)
        m_surface = re.search(r'h(\d+)_surface', col)
        if m_depth:
            hole, depth = int(m_depth.group(1)), int(m_depth.group(2))
            return (hole, -depth, 0) 
        elif m_surface:
            hole = int(m_surface.group(1))
            return (hole, 1, 1)  
        else:
            return (99, 0, 0)
    return sorted(cols, key=sort_key)

def rename_vwc_columns(df: pd.DataFrame, slimeid: str) -> pd.DataFrame:
    site_code = slimeid_map.get(slimeid)
    mapping = selb_map if site_code == 'SELB' else malmms_map
    rename_dict = {}
    for col in df.columns:
        match = re.search(r'VWCm\((\d+)\)', col)
        if match:
            ch = match.group(1)
            if ch in mapping:
                rename_dict[col] = f'soilvwc_{mapping[ch]}_m3m3'
    return df.rename(columns=rename_dict)

def rename_uwt_columns(df: pd.DataFrame, slimeid: str) -> pd.DataFrame:
    prefix = slimeid[:2].lower()
    rename_dict = {}
    pattern = re.compile(rf'{prefix}exp_TC\d+_(-?\d+)cm_(above|below|shoreline)', re.IGNORECASE)
    for col in df.columns:
        match = pattern.search(col)
        if match:
            depth, loc = match.groups()
            rename_dict[col] = f'uwtemp_{depth}cm_{loc.lower()}_degc'
    return df.rename(columns=rename_dict)

def sort_uwtemp_columns(cols: list) -> list:
    def sort_key(col):
        m = re.search(r'uwtemp_(-?\d+)cm_(above|below|shoreline)_degc', col)
        if m:
            depth = int(m.group(1))
            # Sort below < shoreline < above for same depth if needed
            loc = m.group(2)
            loc_order = {'below': 0, 'shoreline': 1, 'above': 2}
            return (depth, loc_order.get(loc, 99))
        return (999, 99)
    
    return sorted(cols, key=sort_key)
    
def clean_sensor_failures(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()
    bad_values = [7999, -7999, 9999, -9999, '7999', '-7999', '9999', '-9999', '-INF', 'INF','NAN']
    df.replace(bad_values, np.nan, inplace=True)
    df = df.replace([np.inf, -np.inf], np.nan)
    for col in df.select_dtypes(include=[np.number]).columns:
        df[col] = df[col].astype(float)
    return df

bur_renamed = rename_ecp_columns(bur, slimeid)
bur_renamed = rename_temp_columns(bur_renamed, slimeid)
bur_renamed = rename_vwc_columns(bur_renamed, slimeid)
exp_renamed = rename_uwt_columns(exp, slimeid)
soilec_cols = [c for c in bur_renamed.columns if c.startswith('soilec_')]
soilt_cols = [c for c in bur_renamed.columns if c.startswith('soilt_')]
soilvwc_cols = [c for c in bur_renamed.columns if c.startswith('soilvwc_')]
uwtemp_cols = [c for c in exp_renamed.columns if c.startswith('uwtemp_')]
soilec_cols_sorted = sort_by_hole_depth(soilec_cols)
soilt_cols_sorted = sort_by_hole_depth(soilt_cols)
soilvwc_cols_sorted = sort_by_hole_depth(soilvwc_cols)
uwtemp_cols_sorted = sort_uwtemp_columns(uwtemp_cols)
```

```python
subm_file = pd.DataFrame()
subm_file['dataset_code'] = [f'almms_{slimeid_lc}'] * len(bur)
subm_file['date_time'] = exp['TIMESTAMP']
subm_file['airt_degc'] = exp['T107_C']
if slimeid in ["SELB", "NELB"]:
    par_col = f"{slimeid}_UW_PAR_uncorr_Avg"
else:
    par_col = "UW_PAR_uncorr_Avg"
subm_file['uwpar_umolsm2'] = exp[par_col]
for col in soilec_cols_sorted:
    subm_file[col] = bur_renamed[col]
for col in soilt_cols_sorted:
    subm_file[col] = bur_renamed[col]
for col in soilvwc_cols_sorted:
    subm_file[col] = bur_renamed[col]
for col in uwtemp_cols_sorted:
    subm_file[col] = exp_renamed[col]
subm_file['do_pct'] = exp['DO_sat']
subm_file = clean_sensor_failures(subm_file)
subm_file
subm_file.to_csv(f'{downloads_path}/{slimeid}_{season}_ALMMS_SUBM.csv', index = False,na_rep = '')
```
